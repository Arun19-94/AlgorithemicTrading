```python
time_period = 20 # number of days over which to average
history = [] # to track a history of prices
sma_values = [] # to track simple moving average values
bank= pd.DataFrame()
banks = [  "SBIN" ,  "HDFCBANK"]
for  t in banks:
    print(t)
    bank[t] =  yf.Ticker(str(t)+'.NS').history(period="40d")["Close"]
    bank["SMC_"+ str(t)] = None
bank
```

```python
for  t in banks:
    print(t)
    bank["SMC_"+ str(t)] =bank[t].rolling(window=time_period).mean()
bank
```
